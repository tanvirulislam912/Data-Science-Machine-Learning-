```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Dhaka Rent.csv')
```

```python
df.head()
```

```python
x = df['area']
y = df['rent']
```

# step 01

```python
m = 0
c = 0
l = 0.0001
```

```python
from IPython.display import Image
Image('Step 02.png')
```

# Next step 

```python
n = len(x)
itr = 1000

for i in range(itr):
    predy = m*x + c
    Dm = (-2/n) * sum(x*(y-predy))
    Dc = (-2/n) * sum(y-predy)
    
    
    #update
    m = m - l*Dm
    c = c - l*Dc
    print(m, c)
    
    
    
```
