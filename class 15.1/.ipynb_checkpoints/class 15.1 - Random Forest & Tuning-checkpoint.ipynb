# Apply Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(xtrain,ytrain)                      #train the model with 70% of data from dataset
pred = dt.predict(xtest)                   # pred is corrsponding to ytest
```

# Evaluate the ML model

```python
from sklearn.metrics import classification_report , accuracy_score, confusion_matrix
classification_report(ytest,pred) 
print(classification_report(ytest,pred))

test_confusion_matrix = confusion_matrix(ytest,pred)
print("--------------------------------------------")
print(test_confusion_matrix)
```

# Grid Search CV

```python
rfc=RandomForestClassifier(random_state=42)
```

```python
# creating param_distributions for random forest
import numpy as np
trees = np.random.randint(15,150,30)    # start,end,no of value
```

```python
max_features = ['sqrt', 'log2']
max_depth = np.random.randint(2,12,50)
criterion = ['gini', 'entropy']
min_samples_split = [2,3,4]


grids ={
    'n_estimators' : trees,
    'max_features' : max_features,
    'max_depth' : max_depth,
    'criterion' : criterion,
    'min_samples_split' : min_samples_split
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
CV_rfc =  GridSearchCV(cv=5, estimator=RandomForestClassifier(random_state=42),
             param_grid={'criterion': ['gini', 'entropy'],
                         'max_depth': [4, 5, 6, 7, 8],
                         'max_features': ['auto', 'sqrt', 'log2'],
                         'n_estimators': [200, 500]})



```

```python
CV_rfc.fit(xtrain, ytrain)
```

```python
CV_rfc.best_params_
```

```python
rfc1=RandomForestClassifier(random_state=42, max_features='auto', n_estimators= 200, max_depth=8, criterion='gini')
```

```python
rfc1.fit(xtrain, ytrain)
```

```python
pred=rfc1.predict(xtest)
```

```python
print("Accuracy for Random Forest on CV data: ",accuracy_score(ytest,pred)*100)
```

# Randomized Search CV

```python
# creating param_distributions for random forest
import numpy as np
trees = np.random.randint(15,150,30)
```

```python
trees
```

```python
max_features = ['sqrt','log2']
max_depth = np.random.randint(2,12,50)
criterion = ['gini','entropy']
min_samples_split = [2,3,4]

grids = {
    'n_estimators' : trees,
    'max_features' : max_features,
    'max_depth' : max_depth,
    'criterion' : criterion,
    'min_samples_split' : min_samples_split
}
```

```python
print(grids)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
rcv = RandomizedSearchCV(clf, grids, n_iter=200, cv=3)
```

```python
rcv.fit(xtrain, ytrain)
```

```python
rcv.cv_results_
```

```python
tuning_result = pd.DataFrame(rcv.cv_results_)
```

```python
tuning_result
```

```python
rcv.best_score_
```

```python
rcv.best_params_
```

```python
clf2 = rcv.best_estimator_
```

```python
clf2.fit(xtrain, ytrain)   # training again
```

```python
clf2.score(xtest, ytest)*100
```
