# Model Implementation

```python
df2 = df1.copy()
```

```python
X = df1.drop('diagnosis', axis=1).values
Y = df1['diagnosis'].values

```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split (X, Y, test_size = 0.30, random_state=1)
```

# Algorithm Implementation

# Using DecisionTreeClassifier

```python
#Using DecisionTreeClassifier 
from sklearn.tree import DecisionTreeClassifier
trees = DecisionTreeClassifier(criterion = 'entropy', random_state = 1)
trees.fit(X_train, Y_train)
trees.predict(X_test)
```

```python
np.array(Y_test)
```

```python
score = trees.score(X_train, Y_train)*100
score
```

```python
score = trees.score(X_test, Y_test)*100
score
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(trees.predict(X_test), Y_test)*100
accuracy
```

```python
trees.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(Y_test,trees.predict(X_test) ))
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
print("MSE= ", mean_squared_error(Y_test,trees.predict(X_test))*100) # The mean squared error
print("MAE =", mean_absolute_error(Y_test, trees.predict(X_test)*100)) # The mean absolute error
```

# Using KNeighborsClassifier

```python
#Using KNeighborsClassifier 
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
knn.fit(X_train, Y_train)
knn.predict(X_test)
```

```python
np.array(Y_test)
```

```python
score = knn.score(X_train, Y_train)*100
score
```

```python
score = knn.score(X_test, Y_test)*100
score
```

```python
accuracy = accuracy_score(knn.predict(X_test), Y_test)*100
accuracy
```

```python
print(classification_report(Y_test,knn.predict(X_test) ))
```

```python
print("MSE = ", mean_squared_error(Y_test,knn.predict(X_test))*100) # The mean squared error
print("MAE =", mean_absolute_error(Y_test, knn.predict(X_test)*100)) # The mean absolute error
```
