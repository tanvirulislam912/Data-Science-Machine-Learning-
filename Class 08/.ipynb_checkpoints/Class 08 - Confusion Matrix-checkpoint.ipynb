```python
x = df.drop('DEATH_EVENT', axis=1)
```

```python
y = df['DEATH_EVENT']
```

```python
from sklearn.model_selection import train_test_split 
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=.30, random_state=1)           #train_size=.70
```

```python
xtrain.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(xtrain, ytrain)
```

```python
pred = clf.predict(xtest)              #predicted y by testing x value
pred
```

```python
len(pred)
```

```python
import numpy as np                 #actual ytest
np.array(ytest)
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, roc_curve, plot_roc_curve
```

```python
print(classification_report(ytest, pred))
```

```python
cm =  confusion_matrix(ytest, pred)
```

```python
cm
```

```python
clf.score(xtest, ytest)
```

```python
 accuracy_score(ytest, pred)
```

```python
tp, fn, fp, tn =   confusion_matrix(ytest, pred).reshape(-1)
```

```python
tp
```

```python
tn
```

```python
fp
```

```python
(tp+tn) / (tp+tn+fp+fn)
```
