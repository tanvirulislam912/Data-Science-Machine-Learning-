# Ordinal Encoder

```python
df7.head()
```

```python
df7.smoker.unique()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
response=['yes', 'no']
```

```python
ordinal = OrdinalEncoder(categories=[response])
encoded = ordinal.fit_transform(df7[['smoker']])
```

```python
encoded
```

```python
new_Smoker = pd.DataFrame(encoded, columns=['smoker'])
```

```python
new_Smoker.head()
```

```python
df7.head()
```

```python
df7.smoker = new_Smoker
```

```python
df7.head()
```

```python
cols = ['smoker']
for col in cols:
    unique = df8[col].unique()
    df8[col] = OrdinalEncoder(categories=[unique]).fit_transform(df8[[col]])
```
